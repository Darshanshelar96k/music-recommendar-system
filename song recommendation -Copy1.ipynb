```python
df_final = df.drop(columns={'Unnamed: 0','song_ID','artists','danceability','energy','valence','acousticness','loudness','mode','speechiness','tempo','instrumentalness','meditation/relaxation music'})
```

```python
df_final
```

```python
# converting long data into wide data using pivot table
user_music_df=df_final.pivot_table(index='user_ID',columns='song_name',values='rating').reset_index(drop=True)
user_music_df
```

```python
user_music_df.index = df_final.user_ID.unique()
user_music_df.fillna(0,inplace=True)
user_music_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_music_df.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = df_final.user_ID.unique()
user_sim_df.columns = df_final.user_ID.unique()
```

```python
user_sim_df
```

```python
import difflib
```

```python
music_name = input("Enter music you want listen :  ")
```

```python
list_of_all_music_name = df_final['song_name'].tolist()
list_of_all_music_name
```

```python
find_close_match = difflib.get_close_matches(music_name,list_of_all_music_name)
close_match = find_close_match[0]
```

```python
index_of_the_music_name = df_final[df_final.song_name == close_match]['user_ID'].values[0]
index_of_the_music_name
```

```python
user_sim_score = list(enumerate(user_sim_df[index_of_the_music_name]))
user_sim_score
```

```python
sorted_simalar_music = sorted(user_sim_score,key = lambda x:x[1],reverse= True)
print("suggested music : \n")
i = 1
for music in sorted_simalar_music:
    index = music[0]
    name_from_index = df_final[df_final.index == index]['song_name'].values[0]
    if(i<20): 
        print(i,',',name_from_index)
        i+=1
```

# recommendation system for music

```python
music_name = input("Enter music you want listen :  ")
list_of_all_music_name = df_final['song_name'].tolist()
find_close_match = difflib.get_close_matches(music_name,list_of_all_music_name)
close_match = find_close_match[0]
index_of_the_music_name = df_final[df_final.song_name == close_match]['user_ID'].values[0]
user_sim_score = list(enumerate(user_sim_df[index_of_the_music_name]))
sorted_simalar_music = sorted(user_sim_score,key = lambda x:x[1],reverse= True)
print("suggested music : \n")
i = 1
for music in sorted_simalar_music:
    index = music[0]
    name_from_index = df_final[df_final.index == index]['song_name'].values[0]
    if(i<20): 
        print(i,',',name_from_index)
        i+=1
```
